```python
from kafka import KafkaProducer
from time import sleep
import json
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("./datasets/nyc_taxi.csv", parse_dates=["tpep_pickup_datetime", "tpep_dropoff_datetime"])
df = df[["tpep_pickup_datetime", "pickup_x", "pickup_y"]]
df = df.rename(columns={
    'tpep_pickup_datetime': 'ts', 
    'pickup_x': 'longitude',
    'pickup_y': 'latitude'
})
df['longitude'] = df['longitude']/10**5
df['latitude'] = df['latitude']/10**5
df.head(5)
```

```python
import time
producer = KafkaProducer(bootstrap_servers=["host:9092"])
N = 100
# len(df) # 10 million records
for i in range(len(df)):
#     df.at["ts", i] = int(time.now())
    data = f'{int(time.time())},{df.iloc[i]["latitude"]},{df.iloc[i]["longitude"]}'.encode()
#     producer.send("nyc_trips", df.iloc[i].to_json().encode())
    producer.send("nyc_trips", data)
    print(f"sent {i}")
    if (i%N == 0):
        sleep(1)
```
